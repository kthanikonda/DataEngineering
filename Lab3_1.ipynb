```python
list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    clean3 = re.sub(r'[\r\n]+', ' ', clean2)
    clean3 = ' '.join(clean3.split())

    list_rows.append(clean3)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

```python
df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
col_labels = soup.find_all('th')
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df5.info()
df5.shape
```

```python
df6 = df5.dropna(axis=0, how='any')
df6.info()
df6.shape
```

```python
df7 = df6.drop(df6.index[0])
df7.head()
```

```python
df7.rename(columns={'[Place': 'Place'},inplace=True)
df7.rename(columns={' Age Group Place]': 'Age Group Place'},inplace=True)
df7.head()
```

```python
df7['Age Group Place'] = df7['Age Group Place'].str.strip(']')
df7.head()
```

```python
df7.columns = df7.columns.str.strip()

time_list = df7['Time'].tolist()

time_mins = []
for i in time_list:
    parts = i.strip().split(':')
    if len(parts) == 3:
        h, m, s = parts
        minutes = int(h)*60 + int(m) + int(s)/60
    elif len(parts) == 2:
        m, s = parts
        minutes = int(m) + int(s)/60
    else:
        minutes = None
    time_mins.append(minutes)


df7['Runner_mins'] = time_mins


df7.head()
```

```python
df7.describe(include=[np.number])
```

```python
from pylab import rcParams

rcParams['figure.figsize'] = 15, 5

df7.boxplot(column='Runner_mins')
plt.grid(True, axis='y')
plt.ylabel('Time')
plt.xticks([1], ['Runners'])
```

```python
x = df7['Runner_mins']
ax = sns.distplot(x, hist=True, kde=True, rug=False, color='m', bins=25, hist_kws={'edgecolor':'black'})
plt.show()
```

```python
f_fuko = df7.loc[df7['Gender']==' F']['Runner_mins']
m_fuko = df7.loc[df7['Gender']==' M']['Runner_mins']
sns.distplot(f_fuko, hist=True, kde=True, rug=False, hist_kws={'edgecolor':'black'}, label='Female')
sns.distplot(m_fuko, hist=False, kde=True, rug=False, hist_kws={'edgecolor':'black'}, label='Male')
plt.legend()
```
